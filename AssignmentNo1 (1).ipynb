```python
Eng = float(input("Enter your English Maeks: "))
Maths = float(input("Enter your Maths Maeks: "))
Isl = float(input("Enter your Isl Maeks: "))
Urdu = float(input("Enter your Urdu Maeks: "))
Science = float(input("Enter your Science Maeks: "))

Total_Marks = Eng + Isl + Maths + Science + Urdu

precentage = (Total_Marks/500) * 100

if percentage >= 80 and percentage <=100:
      print("A-1 Grade")
elif percentage >= 70 and percentage < 80:
      print("A Grade")
elif percentage >= 60 and percentage < 70:
      print("B Grade")
elif percentage >= 50 and percentage < 60:
      print("C Grade")
elif percentage >= 40 and percentage < 50:
      print("D Grade")
elif percentage >= 33 and percentage < 40:
      print("E Grade")
else:
      print("Fail")
```
