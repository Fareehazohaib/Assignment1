```python
dictionary = {
    "a" : 10,
    "b" : 20,
    "c" : "Hello",
    "d" : 30,
}

total = 0
for values in dictionary:
    if type(values) == int or float:
     total += values
print("Total sum of numeric values: ", total)
```
